```python
# NAMA  : JIHAD NATRA SITUMORANG
# NIM   : 1207070061
# KELAS : PCD TSEB
# ======================================================================================================================

# Import dependency
import numpy as np
import imageio.v2
import matplotlib.pyplot as plt

# Membaca gambar
img = imageio.v2.imread("D:\Belajar Python\Pertemuan 9\1207070061_Jihad Natra  S_PRAKTIKUM 5\image\Kocheng.jpg")

# Mendapatkan resolusi dan type dari gambar
img_height = img.shape[0]# Mengambil tinggi gambar
img_width = img.shape[1]# Mengambil lebar gambar
img_channel = img.shape[2]# Mengambil jumlah channel gambar
img_type = img.dtype# Mengambil tipe data gambar
# ======================================================================================================================
# Contrass Autolevel
# ======================================================================================================================

# Membuat variabel img_contrass untuk menampung hasil
img_autocontrass = np.zeros(img.shape, dtype=np.uint8)

# Melakukan penambahan contrass secara otomatis
def autocontrass():
    xmax = 300
    xmin = 0
    d = 0
    # Mendapatkan nilai d, dimana nilai d ini akan berpengaruh pada perhitungan
    # untuk mendapatkan tingkat kontras

    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            # Menghitung tingkat keabuan piksel dengan rata-rata komponen warna

            if gray < xmax:
                xmax = gray
            # Memperbarui nilai maksimum (xmax) jika tingkat keabuan piksel lebih kecil dari xmax

            if gray > xmin:
                xmin = gray
            # Memperbarui nilai minimum (xmin) jika tingkat keabuan piksel lebih besar dari xmin

    d = xmin - xmax
    # Menghitung d sebagai selisih antara nilai minimum dan maksimum

    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            # Menghitung tingkat keabuan piksel dengan rata-rata komponen warna

            gray = int(float(255 / d) * (gray - xmax))
            # Menghitung tingkat keabuan yang telah disesuaikan dengan menggunakan rumus kontras autolevel

            img_autocontrass[y][x] = (gray, gray, gray)
            # Menyimpan tingkat keabuan yang telah disesuaikan ke dalam gambar "img_autocontrass" dengan menggunakan tiga komponen warna yang sama (gray, gray, gray)


# Menampilkan hasil autolevel contrass
autocontrass()
# Memanggil fungsi autocontrass untuk melakukan penyesuaian kontras otomatis pada gambar

plt.imshow(img_autocontrass)
# Menampilkan gambar hasil penyesuaian kontras otomatis

plt.title("Contrass Autolevel")
# Memberikan judul pada plot dengan teks "Contrass Autolevel"

plt.show()
# Menampilkan plot gambar dengan kontras otomatis
```
